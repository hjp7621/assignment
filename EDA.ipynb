```python
'''
데이터 분석의 종류:
기술통계학 (Descriptive Statistics): 데이터를 요약, 정리, 설명하는 통계적 방법을 다룹니다. 평균, 중앙값, 표준편차 등이 여기에 속합니다.

탐색적 데이터 분석 (Exploratory Data Analysis - EDA): 데이터의 특성을 이해하고 패턴을 찾는 과정으로, 시각화 및 기술통계를 활용합니다.

추론통계학 (Inferential Statistics): 표본 데이터를 사용하여 모집단에 대한 결론을 내리는 통계 방법입니다. 가설검정과 신뢰구간이 여기에 속합니다.

회귀분석 (Regression Analysis): 변수 간의 관계를 모델링하고 예측하는 통계 기법입니다.

머신러닝 (Machine Learning): 데이터에서 패턴을 학습하여 예측이나 의사결정을 수행하는 컴퓨터 알고리즘을 사용하는 방법입니다.

비정형 데이터 분석 (Unstructured Data Analysis): 텍스트, 이미지, 음성 등과 같은 구조화되지 않은 데이터를 다루는 분석 방법입니다.

통계의 종류:
기술 통계 (Descriptive Statistics): 데이터를 요약하고 설명하는 통계적 방법입니다.

추리 통계 (Inferential Statistics): 표본 데이터를 사용하여 모집단에 대한 일반화된 결론을 도출하는 통계적 기법입니다.

확률론 (Probability Theory): 불확실성을 다루는 수학적 기초를 제공하며 통계 분석과 밀접한 관련이 있습니다.

다변량 통계 (Multivariate Statistics): 여러 변수 간의 관계를 다루는 통계적 기법입니다.

시계열 분석 (Time Series Analysis): 시간에 따라 변하는 데이터의 패턴을 분석하는 통계적 기법입니다.

생존 분석 (Survival Analysis): 사건이 발생할 때까지의 시간을 다루는 통계 기법으로, 응용 분야로는 의학 및 공학이 있습니다.

베이지안 통계 (Bayesian Statistics): 확률을 믿음의 정도로 해석하고 업데이트하는 방식으로 통계적 추론을 수행하는 방법입니다.
'''


#1. Descriptive Statistics (기술통계학):

import numpy as np

data = np.array([1, 2, 3, 4, 5])

# 평균 계산
mean_value = np.mean(data)

# 중앙값 계산
median_value = np.median(data)

# 표준편차 계산
std_deviation = np.std(data)

print("평균:", mean_value)
print("중앙값:", median_value)
print("표준편차:", std_deviation)



# 2. Inferential Statistics (추론통계학):

from scipy.stats import ttest_ind

# 두 그룹 간의 t-검정 수행
group1 = np.array([1, 2, 3, 4, 5])
group2 = np.array([6, 7, 8, 9, 10])

t_statistic, p_value = ttest_ind(group1, group2)

print("t-통계량:", t_statistic)
print("p-값:", p_value)


# 3. Regression Analysis (회귀분석):

import statsmodels.api as sm

# 예측 변수와 응답 변수 생성
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# 상수항 추가
X = sm.add_constant(X)

# 선형 회귀 모델 적합
model = sm.OLS(y, X).fit()

# 회귀 계수 및 통계량 출력
print(model.summary())


```

```python
# 1. Exploratory Data Analysis (EDA):
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 로드
data = pd.read_csv('your_data.csv')

# 데이터 정보 확인
print(data.info())

# 데이터 통계량 확인
print(data.describe())

# 시각화: 히스토그램
sns.histplot(data['feature'], bins=20)
plt.title('Histogram of Feature')
plt.show()

# 시각화: 산점도 행렬
sns.pairplot(data, hue='target')
plt.title('Pairplot')
plt.show()
```

```python
# 2. p-value 계산:

from scipy.stats import ttest_ind

# 두 그룹의 데이터 생성 (예시)
group1 = data[data['category'] == 'A']['value']
group2 = data[data['category'] == 'B']['value']

# t-test 수행
t_statistic, p_value = ttest_ind(group1, group2)

# 결과 출력
print(f'T-Statistic: {t_statistic}')
print(f'P-Value: {p_value}')
```

```python
# 3. k-means 클러스터링:

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 데이터 전처리
features = data[['feature1', 'feature2']]
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# k-means 모델 학습
kmeans = KMeans(n_clusters=3, random_state=42)
data['cluster'] = kmeans.fit_predict(features_scaled)

# 시각화: 클러스터별 산점도
sns.scatterplot(x='feature1', y='feature2', hue='cluster', data=data)
plt.title('K-means Clustering')
plt.show()
```

# p-value


p-value(유의확률 또는 유의수준)은 통계적 가설 검정에서 사용되는 중요한 개념 중 하나입니다. 가설 검정은 데이터로부터 얻은 결과가 우연에 의한 것인지, 아니면 어떤 실제 효과가 있는지를 판단하는 통계적인 절차입니다. p-value는 이 가설 검정 결과의 통계적 유의성을 나타냅니다.

가설 검정은 크게 다음 두 가지 가설을 검정하는 것으로 이루어집니다:

p-value는 귀무가설이 참이라고 가정할 때, 현재의 표본 데이터나 더 극단적인 결과를 얻을 확률을 나타냅니다. 즉, p-value가 작을수록 귀무가설이 기각되며, 우리는 대립가설을 지지하는 증거가 있다고 해석합니다.

p-value < 유의수준(보통 0.05 또는 0.01): 귀무가설을 기각하고 대립가설을 채택합니다.
p-value ≥ 유의수준: 귀무가설을 기각할 증거가 부족하므로 귀무가설을 기각하지 않습니다.
다음은 p-value에 대한 몇 가지 주의사항입니다:

p-value는 확률을 나타내는 것이 아닙니다: p-value는 주어진 데이터가 귀무가설을 지지하는 정도를 나타내며, 확률 자체를 나타내지 않습니다.
유의수준의 선택: 유의수준은 귀무가설을 기각할 때 얼마나 엄격한지를 결정하는데, 흔히 사용되는 값은 0.05와 0.01입니다.
p-value의 해석은 문맥에 따라 다를 수 있습니다: p-value는 통계적으로 유의하다고 해도, 실제적으로는 중요하지 않을 수 있습니다. 따라서 결과를 실제 상황과 함께 고려해야 합니다.
p-value를 계산하려면 주로 t-test, ANOVA, 회귀분석 등의 통계적인 방법을 사용합니다. Python에서는 scipy.stats 모듈 등을 활용하여 p-value를 계산할 수 있습니다.


# EDA


EDA(Exploratory Data Analysis)는 데이터를 탐색하고 이해하는 과정으로, 데이터의 특성을 파악하고 패턴을 발견하는 데 사용됩니다. EDA를 통해 데이터의 구조, 분포, 이상치, 상관관계 등을 파악할 수 있으며, 이는 효과적인 모델링과 의사결정에 기여합니다. 아래는 EDA를 수행하는 데 사용되는 주요 기법과 단계에 대한 개요입니다:

EDA의 주요 기법:
기술 통계 (Descriptive Statistics):

평균, 중앙값, 표준편차 등의 기술 통계량을 계산하여 데이터의 중심 경향과 분산을 파악합니다.
히스토그램 및 밀도 그림:

데이터의 분포를 시각화하여 데이터의 모양, 이상치, 극단값 등을 확인합니다.
상자 그림 (Box Plot):

데이터의 중앙값, 사분위수, 이상치를 시각적으로 확인할 수 있는 방법입니다.
산점도 행렬 (Pair Plot):

변수들 간의 상관관계를 살펴보기 위해 산점도 행렬을 그립니다.
상관 행렬:

변수들 간의 상관관계를 숫자로 표현한 행렬을 생성하여 시각화합니다.
QQ 플롯 (Quantile-Quantile Plot):

표본 데이터의 분포와 정규 분포의 표본 분포를 비교하여 데이터가 정규 분포를 따르는지 확인합니다.
산점도 및 회귀선:

두 변수 간의 관계를 시각화하여 상관성 및 회귀 관계를 확인합니다.
EDA의 주요 단계:
데이터 로드 및 확인:

데이터를 불러오고 기본적인 정보를 확인합니다.
결측치 및 이상치 처리:

결측치와 이상치를 확인하고 필요한 경우 처리합니다.
기술 통계 및 시각화:

주요 통계량을 계산하고 히스토그램, 상자 그림 등을 이용하여 데이터를 시각화합니다.
변수 간 상관관계 확인:

변수들 간의 상관관계를 파악하여 모델링에 활용할 수 있는 중요한 변수를 선정합니다.
변수 변환:

필요한 경우 변수를 변환하여 분석에 적합한 형태로 만듭니다.
패턴 및 규칙 찾기:

클러스터링, 이상치 탐지 등의 방법을 사용하여 데이터의 패턴과 규칙을 찾습니다.
결론 도출 및 다음 단계 수립:

EDA를 통해 얻은 결과를 종합하여 결론을 도출하고, 다음 분석 단계나 모델링 계획을 수립합니다.
EDA는 데이터 분석 프로세스에서 매우 중요한 초기 단계이며, 데이터 과학 및 머신러닝 프로젝트에서 풍부한 인사이트를 얻기 위해 활용됩니다.






# k-means 클러스터링

k-means 클러스터링은 데이터를 k개의 클러스터로 그룹화하는 비지도 학습 알고리즘입니다. 각 클러스터는 데이터 포인트들 간의 유사성에 기초하여 형성됩니다. k-means 알고리즘은 각 클러스터의 중심(centroid)을 찾아가면서 클러스터를 형성하는 방식으로 작동합니다.

k-means 알고리즘 동작 방식:
클러스터의 개수 선택 (k):

사용자가 몇 개의 클러스터로 데이터를 그룹화할 것인지 결정합니다.
초기 중심점 선택:

각 클러스터의 중심을 초기에 무작위로 선택하거나 특정한 방법을 사용하여 초기화합니다.
각 데이터 포인트를 가장 가까운 중심에 할당:

각 데이터 포인트를 가장 가까운 중심에 할당하여 클러스터를 형성합니다.
각 클러스터의 중심 재계산:

할당된 데이터 포인트를 기반으로 각 클러스터의 중심을 재계산합니다.
수렴 또는 일정 횟수 동안 반복:

할당 및 중심 재계산 단계를 반복하면서 클러스터의 중심이 수렴할 때까지 또는 일정 횟수 동안 반복합니다.
클러스터 할당 완료:

알고리즘이 수렴하면 각 데이터 포인트는 최종적으로 한 클러스터에 속하게 됩니다.
주요 특징:
비지도 학습: 레이블이 없는 데이터에 대해 학습하므로 비지도 학습 알고리즘에 속합니다.

거리 기반: 클러스터링은 데이터 포인트 간의 거리나 유사성을 기반으로 이루어지며, k-means 역시 거리 기반 알고리즘입니다.

클러스터 내의 분산을 최소화하는 원칙: 각 클러스터 내의 데이터 포인트들 간의 분산을 최소화하고, 각 클러스터 간의 중심들 간의 거리를 최대화하는 원칙을 따릅니다.
